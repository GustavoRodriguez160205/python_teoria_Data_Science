# Tipos de Datos en Python

## 1. Números
- **`int`**: Enteros, por ejemplo: `1`, `42`.
- **`float`**: Decimales, por ejemplo: `3.14`.
- **`complex`**: Números complejos, por ejemplo: `2 + 3j`.

## 2. Cadenas
- **`str`**: Texto, por ejemplo: `"Hola"`.

## 3. Booleanos
- **`bool`**: `True` o `False`.

## 4. Listas
- **`list`**: Colecciones mutables, por ejemplo: `[1, 2, 3]`.

## 5. Tuplas
- **`tuple`**: Colecciones inmutables, por ejemplo: `(1, 2, 3)`.

## 6. Conjuntos
- **`set`**: Elementos únicos, por ejemplo: `{1, 2, 3}`.

## 7. Diccionarios
- **`dict`**: Pares clave-valor, por ejemplo: `{'nombre': 'Juan'}`.

## 8. None
- **`NoneType`**: Valor nulo, se define como `None`.
